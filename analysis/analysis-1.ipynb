```python
#find total with hearing difficulty in poverty 
#read clean csv
dfv = pd.read_csv("data/disability-poverty.csv")
dfv.head(10)
```

```python
#filter under .5
dfs1 = dfv.query('Ratio == "Under .50:"')
dfs1
```

```python
#filter .5 to .99
dfs2 = dfv.query('Ratio == ".50 to .99:"')
dfs2
```

```python
#combine filtered sets
#there had to be a way to do both these things but idk it
dfs = pd.concat([dfs1, dfs2], ignore_index = True)
dfs
```

```python
#filter for vision difficulty 
dfs_v = dfs.query('Type == "With a vision difficulty"')
dfs_v
```

```python
#sum for vision
v_sum = dfs_v["Pop"].sum()
v_sum 
```

```python
print(f"In New York state, there are {v_sum} people with vision difficulties living under the poverty line.")
```

```python
#filter for hearing difficulty 
dfs_h = dfs.query('Type == "With a hearing difficulty"')
dfs_h
```

```python
#sum for hearing
h_sum = dfs_h["Pop"].sum()
h_sum 
```

```python
print(f"In New York state, there are {v_sum} people with vision difficulties and {h_sum} people with hearing difficulties living under the poverty line.")
```
